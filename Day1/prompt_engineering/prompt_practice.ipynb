```python
from openai import OpenAI
import os


api_key= os.environ.get("OPENAI_API_KEY")

client = OpenAI(
    api_key=os.environ.get("OPENAI_API_KEY"),
)


def call_gpt(system_prompt: str, user_prompt: str):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt},
        ]
    )
    
    return response.choices[0].message.content
```

```python
prompt_system_math = "당신은 유능한 수학자 입니다."
prompt_system_ml = "당신은 유능한 머신러닝 엔지니어 입니다."
prompt_user_query = "벡터에 대해서 알려주세요. 벡터 사용의 응용 분야 예제 1가지에 대해서 상세히 설명해 주세요."

print(call_gpt(prompt_system_math, prompt_user_query))
print("###########")
print(call_gpt(prompt_system_ml, prompt_user_query))
```

```python
prompt_user_query2 = "벡터에 대해서 알려주세요. 벡터 사용의 응용 분야 예제 1가지에 대해서 상세히 설명해 주세요. step by step 으로 단계별로 설명해주세요."

print(call_gpt(prompt_system_math, prompt_user_query2))
print("###########")
print(call_gpt(prompt_system_ml, prompt_user_query2))
```
